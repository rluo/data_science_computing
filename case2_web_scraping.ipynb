# Download PUMS data (low-level coding)
Go to https://www2.census.gov/programs-surveys/acs/data/pums/2018/5-Year/

```python
from urllib.request import urlopen, urlretrieve
import time
```

```python
url = 'https://www2.census.gov/programs-surveys/acs/data/pums/2018/5-Year/'
contents = urlopen(url).read().decode('utf-8').split('\n')
```

```python
contents
```

```python
csv_names = [line[84:95] for line in contents if 'csv' in line]
```

```python
csv_names
```

```python
# https://www2.census.gov/programs-surveys/acs/data/pums/2018/5-Year/csv_hak.zip
```

```python
# Download the files automatically
for csv_name in csv_names:
    print("Downloading %s" % csv_name)
    urlretrieve(url + csv_name, csv_name)
    # To avoid overwhelming the web traffic,
    # add pauses between downloads to human clicking.
    time.sleep(10)
```

# Extract Wiki Tables (medium level coding)

```python
import urllib.request
from bs4 import BeautifulSoup

url = "https://en.wikipedia.org/wiki/S%26P_100"
print(url)
```

```python
with urllib.request.urlopen(url) as resp:
    html = resp.read()
```

```python
print(html)
```

```python
soup = BeautifulSoup(html, "lxml")
```

```python
soup.findAll("table")
```

```python
tab = soup.findAll("table")[2]
print(tab)
```

```python
for row in tab.findAll("tr"):
    print( [col.text  for col in row.findAll("td") ] )
```

# Download Yahoo Stock Data (high level coding)

```python
from urllib.request import urlretrieve
import time
import random
from selenium import webdriver
from selenium.webdriver.common.by import By

# download driver from https://chromedriver.chromium.org/downloads
```

```python
# Try the standard way for downloading webpages
symbol = 'XOM'
url = 'https://finance.yahoo.com/quote/XOM/history?p=' + symbol
urlretrieve(url, 'output2.html')
```

```python
# Selenium
chrome_options = webdriver.ChromeOptions()
chrome_options.add_experimental_option('prefs', {
    "profile.default_content_settings.popups": 0,
    "download.prompt_for_download": False,
    "download.default_directory": "/Users/xluo/Downloads/"
})
driver = webdriver.Chrome(options=chrome_options)


driver.get("https://finance.yahoo.com/quote/XOM/history?p=XOM")
time.sleep(10)
```

```python
hp_csv = driver.find_element(By.XPATH, "//a[contains(@download, 'csv')]")
```

```python
hp_csv.click()
```

```python
driver.close()
```
